```python
# Description and identification for the dataset from 'Kaggle':

"""
Name: Description
Year: 2008
Month: 1-12
DayofMonth: 1-31
DayOfWeek: 1 (Monday) - 7 (Sunday)
DepTime: actual departure time (local, hhmm)
CRSDepTime: scheduled departure time (local, hhmm)
ArrTime: actual arrival time (local, hhmm)
CRSArrTime: scheduled arrival time (local, hhmm)
UniqueCarrier: unique carrier code
FlightNum: flight number
TailNum: plane tail number
ActualElapsedTime: in minutes
CRSElapsedTime: in minutes
AirTime: in minutes
ArrDelay: arrival delay, in minutes
DepDelay: departure delay, in minutes
Origin: origin IATA airport code
Dest: destination IATA airport code
Distance: in miles
TaxiIn: taxi in time, in minutes
TaxiOut: taxi out time in minutes
Cancelled: was the flight cancelled?
CancellationCode: reason for cancellation (A = carrier, B = weather, C = NAS, D = security)
Diverted: 1 = yes, 0 = no
CarrierDelay: in minutes
WeatherDelay: in minutes
NASDelay: in minutes
SecurityDelay: in minutes
LateAircraftDelay: in minutes
"""
```

```python
data.reset_index(inplace=True)
```

```python
# Dropping useless columns :
data.drop(columns=['index', 'TaxiIn', 'TaxiOut', 'Year', 'Unnamed: 0', 'DayofMonth', 'FlightNum', 'TailNum', 'CancellationCode', 'Cancelled', 'Origin', 'Dest', 'Diverted'], axis=1, inplace=True)
```

```python
data.head()
```

```python
# Creating a feauture colled Season by checking the Month value
# This feature is import vue the change of the weather

data['Season'] = ['Winter' if x == 12 or x == 1 or x == 2 else 'Spring' if x == 3 or x == 4 or x == 5 else 'Summer' if x == 6 or x == 7 or x == 8 else 'Automn' for x in data.Month]
data.head()
```

```python
data.columns
```

```python
# Changing the position of the 'Season' feature to see it behind the 'Month' feature

data = data[['Month', 'Season', 'DayOfWeek', 'DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime',
       'UniqueCarrier', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime',
       'ArrDelay', 'DepDelay', 'Distance', 'CarrierDelay', 'WeatherDelay',
       'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
```

```python
data.head()
```

```python
DayOfWeek = {1 : 'Mon',
             2 : 'Tues',
             3 : 'Wed',
             4 : 'Thur',
             5 : 'Fri',
             6 : 'Sat',
             7 : 'Sun'}

data.DayOfWeek = data.DayOfWeek.apply(lambda x : DayOfWeek[x])
```

```python
# With the 'UniqueCarrier' feature we can create another feauture called 'AirLineCompany' by Googling !

UniqueCarrier = {'WN' : 'Southwest Airlines Co.',
                 'AA' : 'American Airlines Inc.',
                 'MQ' : 'American Eagle Inc.',
                 'UA' : 'United Airlines, Inc.',
                 'OO' : 'SkyWest  Airlines',
                 'DL' : 'Delta Air Lines Inc.',
                 'XE' : 'ExpressJet Airlines, Inc.',
                 'CO' : 'Continental Airlines, Inc.',
                 'US' : 'US Airways, Inc.',
                 'EV' : 'Atlantic Southeast Airlines, Inc.',
                 'NW' : 'Northwest Airlines, Inc.',
                 'FL' : 'AirTran Airways, Inc.',
                 'YV' : 'Mesa Airlines, Inc.',
                 'B6' : 'JetBlue Airways Inc.',
                 'OH' : 'Comair, Inc.',
                 '9E' : 'Pinnacle Airlines Inc.',
                 'AS' : 'Alaska Airlines Inc.',
                 'F9' : 'Frontier Airlines, Inc.',
                 'HA' : 'Hawaiian Airlines, Inc.',
                 'AQ' : 'Aloha Air Inc.'}

data['AirLineCompany'] = data.UniqueCarrier.apply(lambda x : UniqueCarrier[x])
```

```python
#Changing the position of the 'AirLineCompany' to place it behind the 'UniqueCarrier' feature

data = data[['Month', 'Season', 'DayOfWeek', 'DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime',
       'UniqueCarrier', 'AirLineCompany', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime',
       'ArrDelay', 'DepDelay', 'Distance', 'CarrierDelay', 'WeatherDelay',
       'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
data.head()
```

```python
# Now we're going to bin the 'Distance' feature

binwidth = (max(data['Distance'])-min(data['Distance']))/4
bins = np.arange(min(data['Distance']), max(data['Distance']), binwidth)
group_names = ['1', '2', '3']
# 1 means Close, 2 means Medium_Distance, 3 means Far
data['Distance_Binned'] = pd.cut(data['Distance'], bins=bins, labels=group_names)
data['Distance_Binned'] = data['Distance_Binned'].astype('float')

#Changing the position of the 'Distance_Binned' to place it behind the 'Distance' feature

data = data[['Month', 'Season', 'DayOfWeek', 'DepTime', 'CRSDepTime', 'ArrTime', 'CRSArrTime',
       'UniqueCarrier', 'AirLineCompany', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime',
       'ArrDelay', 'DepDelay', 'Distance', 'Distance_Binned', 'CarrierDelay', 'WeatherDelay',
       'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
data.head()
```

```python
# We can note that 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay' have some few null values
# So let's drop it :

data.dropna(inplace=True)
```

```python
# Creating a version of the data with normalized values and store it in a variable called 'datan' like 'DataNormalized' :


# First We will create a sample of our dataset that selects numeric values and store it in avariable called 'newdf' :

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
datan = data.select_dtypes(include=numerics)

# The second step :

for x in datan.columns:
    datan[x] = datan[x]/datan[x].max()
    
# But also we wouldn't drop the categorical values so we're going to move to the third step :

categorics = ['object']
datan1 = data.select_dtypes(include=categorics)

for y in datan1.columns:
    datan[y] = datan1[y]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Now from Data_Visualization let's move to Modelling :

# We know certainly that Distance depends with the AirTime, ActualElapsedTime, CRSElapsedTime ... :
# So let's use Classification to predict Distance_Binned :
# This model is useful for new passengers who are taking for the first time the plane and our model we're going to announce if the country is close or medium of distance of far :

# Accaling to the topic, kNN algorithm is the best algorithm because we're talking about Distance and everybody knows how does kNN works :

X_train = data.loc[:1734758, ['AirTime']]
y_train = data.loc[:1734758, 'Distance_Binned']

X_test = data.loc[1734760:, ['AirTime']]
y_test = data.loc[1734760:, 'Distance_Binned']

clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train, y_train)

# Now let's predict Distance_Binned given other values :

y_pred = clf.predict(X_test)

# And finally let's calculate the accuracy of the predictions :

print('Accuracy :', accuracy_score(y_test, y_pred)*100, '%')
```

```python
# We saw in the Data_Visualization that the variables we visualize have a very linear relantionship. So, we're going to use the LinearRegression algorithm

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# The first model of regression will be for predicting the Distance between two countries given :
# Now in this ML model we're going to use regression and to predict the Distance itself :

X_train = data.loc[:1734758, ['AirTime']]
y_train = data.loc[:1734758, 'Distance']

X_test = data.loc[1734760:, ['AirTime']]
y_test = data.loc[1734760:, 'Distance']

linreg1 = LinearRegression()
linreg1.fit(X_train, y_train)

# Now we're going to apply the model and predict values

y_pred = linreg1.predict(X_test)

print('r2_score :', r2_score(y_test, y_pred)*100, '%')
```

```python
# The second model of regression will be for predicting The CRSElapsedTime given :

X_train = data.loc[:1734758, ['Distance', 'WeatherDelay']]
y_train = data.loc[:1734758, 'CRSElapsedTime']

X_test = data.loc[1734760:, ['Distance', 'WeatherDelay']]
y_test = data.loc[1734760:, 'CRSElapsedTime']

linreg2 = LinearRegression()
linreg2.fit(X_train, y_train)

# Now we're going to apply the model and predict values

y_pred = linreg2.predict(X_test)

print('r2_score :', r2_score(y_test, y_pred)*100, '%')
```

```python
# The third model of regression intend for for predicting The AirTime given :

X_train = data.loc[:1734758, ['Distance', 'WeatherDelay', 'CarrierDelay', 'NASDelay']]
y_train = data.loc[:1734758, 'AirTime']

X_test = data.loc[1734760:, ['Distance', 'WeatherDelay', 'CarrierDelay', 'NASDelay']]
y_test = data.loc[1734760:, 'AirTime']

linreg3 = LinearRegression()
linreg3.fit(X_train, y_train)

# Now we're going to apply the model and predict values

y_pred = linreg3.predict(X_test)

print('r2_score :', r2_score(y_test, y_pred)*100, '%')
```
